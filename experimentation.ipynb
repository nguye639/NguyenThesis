### Time Series Analysis

```python
graph = Graph()
graph.createRandomComment(0)
sentiment = [getSentimentStatistics(graph)[0]]

for i in range(0,50):
    p = np.random.uniform()
    
    if p < .5 or i == 0:
        graph.addCommenter(0)
    else:
        graph.addReply()
    sentiment.append(getSentimentStatistics(graph)[0])
   
```

```python
plt.figure(figsize = (16,8))

plt.subplot(1,2,1)
graph.plotGraph(showReplies = False)


plt.subplot(1,2,2)
t = np.arange(0,len(sentiment))
plt.plot(t,sentiment)
```
